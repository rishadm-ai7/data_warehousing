Initialisation of required lists and function.

```
fake = Faker()

rand=[]
ph=[]
fid=[]
cred=[]
pay=[]
diseases=["cold","flu","cancer","covid","pneumonia","chickenpox","Diarrhea","stomach ache","AIDS","malaria","dengue"]
condition=["Undetermined","Good","Fair","Serious","Critical"]
services=["Emergency room services","blood test","X-ray/radiology services","Short-term hospitalization","specialty surgical services","general services","Laboratory services"]
doctors=["Dr. Charles Burton","Dr. Mark Hall","Dr. Peter Leavitt","	Dr. Martin Arrowsmith","Dr. Chang","Dr. Rashid","Dr. Grattan","Dr. Edwin Spindrift","Dr. Yuri Zhivago","Dr. Gonzo","Dr. T.J. Eckleburg","Dr. Moreau","Dr. Henry Wu","Dr. Timothy Flyte","Dr. Jennifer Paige"]
```

Creating functions to generate data and unique ID's

```
################function to generate random gender as male or female

def random_gender():
    if (random.randint(0,1)==0):
        return "female"
    else:
        return "male"

################function to generate unique ID in custom format

def gen_id():
    a='%s%s%s%s'%(random.choice(string.ascii_letters.lower()),str(random.randint(1111,9999)),random.choice(string.ascii_letters.lower()),str(random.randint(111,999)))
    
    for x in rand:
        if (a==x):
            return gen_id()
    rand.append(a)
    return a

################function to generate unique phone number

def gen_ph():
    b=random.randint(7000000000,9000000000)
    for x in ph:
        if (b==x):
            return gen_ph()
    ph.append(b)
    return b

################function to generate unique id

def gen_fid():
    c=random.randint(22222,55555)
    for x in fid:
        if (c==x):
            return gen_fid()
    fid.append(c)
    return c

################function to generate unique speciality ID

def gen_spid():
    f=random.choice(["1","2","3","4","5"])
    return f

################function to generate unique credential ID


def gen_credid():
    d=random.randint(7000,9000)
    for x in cred:
        if (d==x):
            return gen_credid()
    cred.append(d)
    return d

################function to generate unique payer ID

def payer_id():
    e=random.randint(5000,7000)
    for x in pay:
        if (e==x):
            return payer_id()
    pay.append(e)
    return e
```

Creating the schema for the src file

```
schema=StructType([
    StructField("Claim_Number",LongType(),True),
    StructField("Claim_Date",DateType(),True),
    StructField("Services_Date_start",DateType(),True),
    StructField("Amount of Charge",IntegerType(),True),
    StructField("Unit of Services",IntegerType(),True),
    StructField("Patient_ID",LongType(),True),
    StructField("Patient_Name",StringType(),True),
    StructField("Patient_Address",StringType(),True),
    StructField("Date_of_birth",DateType(),True),
    StructField("Patient_Phone_Number",LongType(),True),
    StructField("Sex",StringType(),True),
    StructField("Insurance_ID",StringType(),True),
    StructField("Physician_ID",LongType(),True),
    StructField("Physician_Name",StringType(),True),
    StructField("Speciality_ID",StringType(),True),
    StructField("Credential_ID",IntegerType(),True),
    StructField("Review_rating",IntegerType(),True),
    StructField("Diagnosis_ID",IntegerType(),True),
    StructField("Diagnosis",StringType(),True),
    StructField("Patient_Condition",StringType(),True),
    StructField("Patient_Status",StringType(),True),
    StructField("Payer_ID",IntegerType(),True),
    StructField("Payer_Name",StringType(),True),
    StructField("Payer_Address",StringType(),True),
    StructField("Payer_Phone_Number",LongType(),True),
    StructField("Payer_EDI_Number",LongType(),True),
    StructField("Service_Code",IntegerType(),True),
    StructField("Service_Description",StringType(),True),
    StructField("Category_Code",IntegerType(),True),
    StructField("Admitted_Days",IntegerType(),True)
    ])
```

Generating the fake data and appending every row as a list in the list data

```
data=[]
for i in range(1001,1501):
        a=fake.date_between_dates("-10y","-1y")
        data.append([
            int(i),
            fake.date_between_dates(a,"now"),
            a,
            random.randint(5000,25000),
            random.randint(1,10),
            random.randint(777777,999999),
            fake.name(),
            fake.address(),
            fake.date_of_birth(minimum_age=10,maximum_age=90),
            gen_ph(),
            random_gender(),
            gen_id(),
            i+2179,
            random.choice(doctors),
            gen_spid(),
            gen_credid(),
            random.randint(2,5),
            gen_fid(),
            random.choice(diseases),
            random.choice(condition),
            random.choice(["dead","alive"]),
            payer_id(),
            fake.name(),
            fake.address(),
            gen_ph(),
            i+7,
            random.randint(1,50),
            random.choice(services),
            random.randint(0,7),
            random.randint(1,10)
        ])
```
